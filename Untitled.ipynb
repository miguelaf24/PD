```python
#Csv para Dataframe
import csv
import pandas as pd

with open('Foodmart_2020_PD.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    data = list(spamreader)

df = pd.DataFrame(data)
```

```python
for x in range(1,14):

    df[x] = df[x].str.split("=", n = 1, expand = True)
```

```python
import numpy as np

aux = df[1]
for x in range(2,14):
    aux = aux.append(df[x])
```

```python
aux =  list(dict.fromkeys(aux))
aux.remove(None)
```

```python
df_enconded = pd.DataFrame(aux)

def custom():
    
    encoded_vals = []
    for index, row in df.iterrows():
        labels = {}
        uncommons = list(set(aux) - set(row))
        commons = list(set(aux).intersection(row))
        for uc in uncommons:
            labels[uc] = 0
        for com in commons:
            labels[com] = 1
        encoded_vals.append(labels)
    encoded_vals[0]
    ohe_df = pd.DataFrame(encoded_vals)
    
    return ohe_df
ohe_df = custom()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(ohe_df, min_support=0.05, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
rules.head()
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
rules[(rules['confidence'] > 0.2) & (rules['lift'] > 1.0)]
```

#### Compute Frequent Itemsets - Deluxe e Gourmet juntos

```python
deluxe_gourmet_df = deluxe_df.append(gourmet_df).reset_index(drop = True)
def encode_df(df):
    aux = df[1]
    for x in range(2,len(df.columns)):

        aux = aux.append(df[x])

    aux =  list(dict.fromkeys(aux))
    if None in aux: aux.remove(None)

    return pd.DataFrame(aux)
df_encoded_deluxe = encode_df(deluxe_gourmet_df)

def custom(df):
    
    encoded_vals = []
    for index, row in df.iterrows():
        labels = {}
        uncommons = list(set(aux) - set(row))
        commons = list(set(aux).intersection(row))
        for uc in uncommons:
            labels[uc] = 0
        for com in commons:
            labels[com] = 1
        encoded_vals.append(labels)
    encoded_vals[0]
    ohe_df = pd.DataFrame(encoded_vals)
    
    return ohe_df
ohe_deluxe_gourmet = custom(df_encoded_deluxe)
ohe_deluxe_gourmet
```

```python
frequent_itemsets = apriori(ohe_deluxe_gourmet, min_support=0.01, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

#### Regras de Associação

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.05)
rules
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.1)
rules.head()
```
